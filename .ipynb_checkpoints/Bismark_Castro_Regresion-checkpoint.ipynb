```python
%matplotlib notebook
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
%matplotlib inline
import sklearn.datasets as dt
```

```python
def generarDatos():
    '''Se procede a crear los datos ramdomicos de x y y.
           Retornos:
           x[:, 0](Array): Se especifica la filas y columnas del dataset en un array.
           y(Array): Almacena los datos para realizar la regresión lineal.
   '''
    x,y = dt.make_regression(n_samples=100, n_features=1, noise=20, random_state=11)
    return x[:, 0],y
print(generarDatos.__doc__)
```

```python
def calcularRegresion(x, y):
    '''Se realiza el calculo de una regresión lineal de mínimos cuadrados para dos conjuntos de datos.
            Parametros:
            x: Contiene las x del modelo para aprobar crédito.
            y:
            Retornos:
            slope:
            intercept:
            r:
            p:
            std_err:
   '''
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    return slope, intercept, r, p, std_err
print(calcularRegresion.__doc__)
```

```python
'''Se realiza una función auxiliar que contiene un nuevo valor, basicamente evalua los puntos.
            Parametros:
            x: Contiene las x del modelo para aprobar crédito.
            y:
            Retornos:
            slope:
            intercept:
            r:
            p:
            std_err:
   '''
def fitted_value(x):
    return slope * x + intercept
```

```python
def obtenerNuevosValores(x):
  return list(map(fitted_value,x))
```
