```python
%matplotlib notebook
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
import pandas as pd
%matplotlib inline
import sklearn.datasets as dt
```

```python
def generarDatos():
    '''Se procede a crear los datos ramdomicos de x y y.
           Retornos:
           x[:, 0](Array): Se especifica la filas y columnas del dataset en un array.
           y(Array): Almacena los datos para realizar la regresión lineal.
   '''
    x,y = dt.make_regression(n_samples=100, n_features=1, noise=20, random_state=11)
    return x[:, 0],y
print(generarDatos.__doc__)
```

```python
def calcularRegresion(x, y):
    '''Se realiza el calculo de una regresión lineal de mínimos cuadrados para dos conjuntos de datos.
            Parametros:
            x: Contiene las x del modelo.
            y: Contiene las y del modelo.
            Retornos:
            slope: Contiene la pendiente.
            intercept: Contiene la intersección de cualquier llamada de función.
            r: Contiene el radio.
            p: Variable del modelo.
            std_err: Error estándar.
   '''
    slope, intercept, r, p, std_err = stats.linregress(x, y)
    return slope, intercept, r, p, std_err
print(calcularRegresion.__doc__)
```

```python
def fitted_value(x):
    '''Se realiza una función auxiliar que contiene un nuevo valor, basicamente evalua los puntos.
            Parametros:
            x: Contiene las x del modelo.
            Retornos:
            slope: Contiene la pendiente.
            intercept: Contiene la intersección de cualquier llamada de función.
            x: Variable del modelo.
   '''
    return slope * x + intercept
print(fitted_value.__doc__)
```

```python
def obtenerNuevosValores(x):
        '''Se obtiene nuevos valores de la linea de regresión.'''
  return list(map(fitted_value,x))
print(obtenerNuevosValores.__doc__)
```

```python
def graficar(x, y, new_values):
    '''Se dibuja el diagrama de dispersión original para poder dibujar la regresión lineal.
            Parametros:
            x: Contiene las x del modelo.
            y: Contiene las y del modelo.
            new_values:
   '''
    plt.plot(x, y, 'o', label='original data')

    plt.plot(x, new_values, 'r', label='fitted line')

    plt.legend()
print(graficar.__doc__)
```

```python
def main():
    '''Se dibuja el diagrama de dispersión original para poder dibujar la regresión lineal.'''
    x,y = generarDatos()
    global slope
    global intercept
    slope, intercept, r, p, std_err = calcularRegresion(x, y)
    nuevosValores = obtenerNuevosValores(x)
    graficar(x, y, nuevosValores)
print(main.__doc__)
```

```python
main()
```
